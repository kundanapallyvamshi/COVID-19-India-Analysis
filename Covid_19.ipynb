```
india_df['Province/State']=india_df['Province/State'].fillna('others')
print(india_df)
```

```
import matplotlib.pyplot as plt
from matplotlib import *

india_df['ObservationDate']=pd.to_datetime(india_df['ObservationDate'])

date=india_df['ObservationDate']
deaths=india_df['Deaths']

plt.plot_date(date,deaths,linestyle='solid')

plt.gcf().autofmt_xdate()

plt.title('COVID-19 DEATHS')
plt.xlabel('Date')
plt.ylabel('Deaths')

plt.tight_layout()


#print(india_df)

#india_df.plot.hist(subplots=True,grid=True)
```
